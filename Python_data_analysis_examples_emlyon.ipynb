# Data Manipulation

We now move into more advanced territory where we want to heavily manipulate the values in our data. We may want to add our own custom columns, change the type of values or even implement complex functions into our data.

# Apply Function

The apply function is one of the most powerful functions in pandas.
It lets you manipulate the data in whatever form you want to

```python
"""
You have to rename the "Pclass" to their actual names using apply function.
1 = 1st Class
2 = 2nd Class
3 = 3rd Class

"""
```

```python
# Step-1 - define a function
def pclass_name(x):
    if x == 1:
        return "1st Class"
    if x == 2:
        return "2nd Class"
    if x == 3:
        return "3rd Class"
```

```python
df['Pclass'] = df['Pclass'].apply(lambda x : pclass_name(x))
```

```python
df.head(5)
```

# Custom Columns

We can see that the Sex column is an object. For machine learning, we should use numerical values. I want to address 1 for males and 2 for females. 
That's a straightforward task, and you can use the "lambda" function again.

```python
# write your code here

df['Sex'] = df['Sex'].apply(lambda x : 1 if x == "male" else 2)
```
